## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
import json
import pandas as pa
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=r.text

pop_resp=pa.read_json(response)

print(len(pop_resp),pop_resp.columns)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
a=pop_resp[['title','createdAt','commentCount','likeCount']]
a.columns=['標題','貼文時間','留言人數','按讚人數']
print('熱門文章:')
a
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
f=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
fresponse=f.text
npop_resp=pa.read_json(fresponse)

print('熱門文章:')
print('平均留言數',pop_resp['commentCount'].mean())
print('平均按讚人數',pop_resp['likeCount'].mean())
print('非熱門文章:')
print('平均留言數',npop_resp['commentCount'].mean())
print('平均按讚人數',npop_resp['likeCount'].mean())
```
