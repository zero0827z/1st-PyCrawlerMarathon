### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
import xmltodict
import zipfile as zip
f=zip.ZipFile('./Data/example.zip')
f.extractall('./Data/D3')

fh=open('./Data/D3/64_72hr_CH.xml','r',encoding='UTF-8')
xml=fh.read()
fh.close()

d=dict(xmltodict.parse(xml))
da=d['cwbopendata']['dataset']['locations']['location']
for location in da:
    print(location['locationName'])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for i in range(0,38):
    data=d['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]
    print(data['time'][0]['dataTime'],data['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
data=d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]

for i in range(0,24):
    print(data['time'][i]['dataTime'],data['time'][i]['elementValue']['value'])
```
