## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
url='https://www.dcard.tw/_api/forums/pet/posts?popular=true'

r=requests.get(url)
r.encoding='utf-8'
response=r.text
soup=BeautifulSoup(response,'html5lib')
print(soup)
print(type(soup))
```
